#Task 2)Composite Development Score:

```python
# Normalize life expectancy and income (simple scaling)
south_asia_df['life_exp_norm'] = (
    south_asia_df['life_expectancy'] - south_asia_df['life_expectancy'].min()
) / (
    south_asia_df['life_expectancy'].max() - south_asia_df['life_expectancy'].min()
)

south_asia_df['gni_norm'] = (
    south_asia_df['gross_inc_percap'] - south_asia_df['gross_inc_percap'].min()
) / (
    south_asia_df['gross_inc_percap'].max() - south_asia_df['gross_inc_percap'].min()
)

# Composite Score
south_asia_df['Composite Score'] = (
    0.3 * south_asia_df['life_exp_norm'] +
    0.3 * south_asia_df['gni_norm']
)

# Average Composite Score per country (2020–2022)
avg_scores = south_asia_df.groupby('country')['Composite Score'].mean()

# Sorting and selecting top 5
top5 = avg_scores.sort_values(ascending=False).head(5)

print(top5)
```

```python
plt.figure(figsize=(8, 5))
plt.barh(top5.index, top5.values)
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.title("Top 5 South Asian Countries by Composite Score (2020–2022)")
plt.gca().invert_yaxis()
plt.show()
```

#Task 3)OUTLIER DETECTION (IQR)

**Find IQR for HDI and GNI per capita**

```python
hdi_Q1 = south_asia_df['hdi'].quantile(0.25)
hdi_Q3 = south_asia_df['hdi'].quantile(0.75)
hdi_IQR = hdi_Q3 - hdi_Q1

hdi_lower = hdi_Q1 - 1.5 * hdi_IQR
hdi_upper = hdi_Q3 + 1.5 * hdi_IQR


gni_Q1 = south_asia_df['gross_inc_percap'].quantile(0.25)
gni_Q3 = south_asia_df['gross_inc_percap'].quantile(0.75)
gni_IQR = gni_Q3 - gni_Q1

gni_lower = gni_Q1 - 1.5 * gni_IQR
gni_upper = gni_Q3 + 1.5 * gni_IQR
```

**Apply outlier detection**

```python
outliers = south_asia_df[
    (south_asia_df['hdi'] < hdi_lower) | (south_asia_df['hdi'] > hdi_upper) |
    (south_asia_df['gross_inc_percap'] < gni_lower) | (south_asia_df['gross_inc_percap'] > gni_upper)
]
```

```python
outliers_unique = outliers.drop_duplicates(subset='country')
```

**Creating a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.**

```python
plt.figure(figsize=(8,6))
plt.scatter(south_asia_df['gross_inc_percap'], south_asia_df['hdi'], color='blue', label='Normal')
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'], color='red', label='Outlier')
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita with Outliers Highlighted")
plt.legend()
plt.show()

# Step 5: Print only one row per outlier country
print("Outlier countries (one row per country):")
print(outliers_unique[['country', 'hdi', 'gross_inc_percap']])
```

My analysis includes multiple years of data, so each country appears multiple times in the scatter plot. This results in more data points and allows for more reliable outlier detection using the IQR method.

# 7) Outlier Detection:

Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.

country-level averages

```python
# Country-level averages (important!)
sa_avg = south_asia_df.groupby('country')[['hdi', 'gross_inc_percap']].mean()
me_avg = middle_east_df.groupby('country')[['hdi', 'gross_inc_percap']].mean()
```

**Simple function to detect outliers**

```python
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)
```

**Scatter plot with outliers highlighted**

**SOUTH ASIA — HDI vs GNI**

```python
sa_avg['hdi_outlier'] = detect_outliers(sa_avg['hdi'])
sa_avg['gni_outlier'] = detect_outliers(sa_avg['gross_inc_percap'])
sa_avg['outlier'] = sa_avg['hdi_outlier'] | sa_avg['gni_outlier']
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    sa_avg.loc[~sa_avg['outlier'], 'gross_inc_percap'],
    sa_avg.loc[~sa_avg['outlier'], 'hdi'],
    label='Normal',
    alpha=0.7
)

plt.scatter(
    sa_avg.loc[sa_avg['outlier'], 'gross_inc_percap'],
    sa_avg.loc[sa_avg['outlier'], 'hdi'],
    label='Outlier',
    marker='x',
    s=80
)

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('South Asia: HDI vs GNI per Capita (Outliers Highlighted)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

There is a positive association between GNI per capita and HDI in South Asia, indicating that countries with higher income levels generally achieve better human development outcomes. Most observations are concentrated in the lower-to-middle income range, corresponding to moderate HDI values, which reflects the predominantly developing nature of the region. A small number of higher-income observations are identified as outliers using the IQR method, suggesting that income gains in these cases do not translate proportionally into higher HDI. Overall, the pattern indicates that while income growth supports human development, its impact becomes less pronounced at higher income levels.

**MIDDLE EAST — HDI vs GNI**

```python
me_avg['hdi_outlier'] = detect_outliers(me_avg['hdi'])
me_avg['gni_outlier'] = detect_outliers(me_avg['gross_inc_percap'])
me_avg['outlier'] = me_avg['hdi_outlier'] | me_avg['gni_outlier']

plt.figure(figsize=(8,5))

plt.scatter(
    me_avg.loc[~me_avg['outlier'], 'gross_inc_percap'],
    me_avg.loc[~me_avg['outlier'], 'hdi'],
    label='Normal',
    alpha=0.7
)

plt.scatter(
    me_avg.loc[me_avg['outlier'], 'gross_inc_percap'],
    me_avg.loc[me_avg['outlier'], 'hdi'],
    label='Outlier',
    marker='x',
    s=80
)

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('Middle East: HDI vs GNI per Capita (Outliers Highlighted)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
